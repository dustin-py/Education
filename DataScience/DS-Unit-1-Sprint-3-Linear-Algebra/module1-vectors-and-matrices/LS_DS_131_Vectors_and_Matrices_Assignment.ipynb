## 1.1 Create a two-dimensional vector and plot it on a graph

```
# import math and graphing libraries
import matplotlib.pyplot as plt
import numpy as np
```

```
# Create a 2-d vector
two_d = [.1,.2]

# Plot 2-d vector:
fig,ax=plt.subplots()
ax.grid()
plt.xlim(-1,1)
plt.ylim(-1,1)
ax.vlines((0,1),1,-1)
ax.hlines((0,1),1,-1)
plt.arrow(0,0,
          two_d[0],
          two_d[1],
          head_width = .02,
          head_length = .02,
          linewidth = 3,
          color = 'blue',
          length_includes_head=True)

plt.show()
```

## 2.1 What are the dimensions of the following matrices? Which of the following can be multiplied together? See if you can find all of the different legal combinations.
\begin{align}
A = \begin{bmatrix}
1 & 2 \\
3 & 4 \\

\end{bmatrix}
\qquad
B = \begin{bmatrix}
2 & 4 & 6 \\
\end{bmatrix}
\qquad
C = \begin{bmatrix}
9 & 6 & 3 \\
4 & 7 & 11
\end{bmatrix}
\qquad
D = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
\qquad
E = \begin{bmatrix}
1 & 3 \\
5 & 7
\end{bmatrix}
\end{align}

**Dimensions of:**
- A = 3X2
- B = 1X3
- C = 2X3
- D = 3X3
- E = 2X2

```
def dimensions():
    print('Dimension of A: ',A.shape)
    print('Dimension of B: ',B.shape)
    print('Dimension of C: ',C.shape)
    print('Dimension of D: ',D.shape)
    print('Dimension of E: ',E.shape)     
dimensions()                 
```

```
A = np.array([[1,2],[3,4],[5,6]])
B = np.array([[2,4,6]])
C = np.array([[9,6,3],[4,7,11]])
D = np.array([[1,0,0],[0,1,0],[0,0,1]])
E = np.array([[1,3],[5,7]])
```

```
def multiDim(matrix_A,matrix_B):
    combo_list = []
    for i in matrix_A:
        for b in matrix_B:
            if i.shape[1] == b.shape[0]:
                combo_list.append(np.matmul(i,b))
    return combo_list

matrix_A = [A,B,C,D,E]         
matrix_B = [A,B,C,D,E] 
combs = ['AxC','AxE','BxA','BxD','CxA','CxD','DxA','DxD','ExC','ExE'] 
for i in range(0,10):
    print(combs[i],' = \n',multiDim(matrix_A,matrix_B)[i],'\n')
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
CD = np.dot(C,D)
AE = np.dot(A,E)
BA = np.dot(B,A)

print('Original Dimensions:')
dimensions()

print('CD dims: \n',CD.shape)
print('AE dims: \n',AE.shape)
print('BA dims: \n',BA.shape)
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.array([[12,11],
             [7,10]])
J = np.array([[0,1,2],
             [7,10,4],
             [3,2,0]])
print('Determinant of H: \n',np.linalg.det(H),'\n')
print('Determinant of J: \n',np.linalg.det(J),'\n')
```
